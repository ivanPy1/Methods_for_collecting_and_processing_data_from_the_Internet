**1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.**


```python
import requests
from pprint import pprint
import json
```

```python
headers = {
    'User-Agent':
    'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
}
user_name = 'ivanPy1'
url = f'https://api.github.com/users/{user_name}/repos'
```

```python
response = requests.get(url, headers)
```

```python
print(response.status_code)
```

```python
pprint(response.json())
```

```python
response_github_repos = response.json()
```

```python
print(f'Список репозиториев пользователя {user_name}:\n')
for el in response_github_repos:
    print(el['name'])
```

```python
with open(f'list_repos_{user_name}.json', 'w') as f:
    json.dump(response.json(), f)
```

```python
print(f'Список репозиториев пользователя {user_name} в JSON:\n')

for el in response.json():
    print(el['name'])
```
