## Урок 2. Парсинг данных. HTML, DOM, XPath

Написать приложение или функцию, которые собирают основные новости с сайта lenta.ru. Для парсинга использвоать XPath.

Структура данных в виде словаря должна содержать: 

-* название источника;

-наименование новости;

-ссылку на новость;

-дата публикации.

```python
from pprint import pprint
from lxml import html
import requests
import pandas as pd
```

```python
USER_AGENT = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
URL_LENTA = 'https://lenta.ru/parts/news/'
```

```python
header = {
    'User-Agent' : USER_AGENT
}
```

```python
def response_dom (url, header):
    response = requests.get(url, header)
    dom = html.fromstring(response.text)
    
    print(response.status_code, response.url)
    return dom

def dict_data_news (url, dom, source_name_xpath, name_news_xpath, link_news_xpath, publication_date_xpath):
    url_source = 'https://lenta.ru/parts'

    #наименование новости
    name_news = dom.xpath(name_news_xpath)
    
    #ссылка на новость 
    link_news = dom.xpath(link_news_xpath)
    link_news_new = []
    
    for link in link_news:
        if 'https' not in link:
            link = url_source + link
            link_news_new.append(link)
        else:
            link_news_new.append(link)
        
    #дата публикации 
    publication_date = dom.xpath(publication_date_xpath)
    
    #название источника 
    if source_name_xpath == "" :
        source_name = []
        #source_name = ['None' for i in range(len(name_news))]
        for link in link_news:
            if 'https://' in link:
                link = link.replace('https://', '')
                link = link.partition('/')[0]
                source_name.append(link)
            else:
                link = 'lenta.ru'
                source_name.append(link)
    else:
        source_name = dom.xpath(source_name_xpath)
       
    dict_news = {
        "source_name": source_name,
        "name_news": name_news,
        "link_news": link_news_new,
        "publication_date": publication_date        
    }
    
    data_news_df = pd.DataFrame(dict_news)
    
    data_news_df = data_news_df.rename(columns = {'source_name': 'наименование источника',
                                                  'name_news': 'наименование новости',
                                                  'link_news':'ссылка на новость',
                                                  'publication_date':'дата публикации'})
    return data_news_df
```

```python
dict_data_news (url = URL_LENTA,
                dom=response_dom (url=URL_LENTA, header=header),
                source_name_xpath="",
                name_news_xpath="//h3[contains(@class, 'card-full-news__title')]/text()",
                link_news_xpath="//li/a[contains(@class, 'card-full-news _parts-news')]/@href",
                publication_date_xpath="//time[contains(@class, 'card-full-news__info-item card-full-news__date')]/text()"              
               )
```
